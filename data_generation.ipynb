```python
import pandas as pd
import numpy as np
from faker import Faker
import random

fake = Faker()

# Creating Categories DataFrame
categories_data = {
    "CategoryID": [1, 2, 3, 4],
    "CategoryName": ["PVC-U (unplasticised PVC)", "C-PVC (chlorinated PVC)", "PVC-O (molecular oriented PVC)", "Modified PVC"],
    "Description": [fake.text(max_nb_chars=100) for _ in range(4)]
}
categories_df = pd.DataFrame(categories_data)

# Creating Products DataFrame
product_names = ["Agricultural Product", "Plumbing Product", "Industrial Product", "SANITARY PRODUCT", "Water Tanks", "Cleaning Product"]
products_data = {
    "ProductID": list(range(1, 501)),
    "Name": [random.choice(product_names) for _ in range(500)],
    "Description": [fake.text(max_nb_chars=100) for _ in range(500)],
    "CategoryID": [random.choice(categories_data["CategoryID"]) for _ in range(500)],
    "UnitPrice": [round(random.uniform(10, 100), 2) for _ in range(500)],
    "StockQuantity": [random.randint(1, 1000) for _ in range(500)]
}
products_df = pd.DataFrame(products_data)

# Creating Customers DataFrame
customers_data = {
    "CustomerID": list(range(1, 501)),
    "CustomerName": [fake.company() for _ in range(500)],
    "ContactName": [fake.name() for _ in range(500)],
    "Address": [fake.address() for _ in range(500)],
    "Phone": [fake.phone_number() for _ in range(500)],
    "Email": [fake.email() for _ in range(500)]
}
customers_df = pd.DataFrame(customers_data)

# Creating Orders DataFrame
orders_data = {
    "OrderID": list(range(1, 1001)),
    "CustomerID": [random.choice(customers_data["CustomerID"]) for _ in range(1000)],
    "ProductID": [random.choice(products_data["ProductID"]) for _ in range(1000)],
    "OrderDate": [fake.date_between(start_date='-2y', end_date='today') for _ in range(1000)],
    "Quantity": [random.randint(1, 100) for _ in range(1000)],
    "TotalPrice": [round(random.uniform(100, 1000), 2) for _ in range(1000)],
    "Status": [random.choice(["Pending", "Shipped", "Delivered", "Cancelled"]) for _ in range(1000)],
    "SaleID": list(range(1, 1001)),

}
orders_df = pd.DataFrame(orders_data)

# Creating Employees DataFrame
employees_data = {
    "EmployeeID": list(range(1, 101)),
    "EmployeeName": [fake.name() for _ in range(100)],
    "Positions": [random.choice(["Manager", "Salesperson", "Clerk", "Assistant"]) for _ in range(100)],
    "Department": [random.choice(["Sales", "Inventory", "Finance", "HR"]) for _ in range(100)],
    "HireDate": [fake.date_between(start_date='-10y', end_date='today') for _ in range(100)],
    "Phone": [fake.phone_number() for _ in range(100)],
    "CustomerID": [random.choice(customers_data["CustomerID"]) for _ in range(100)],
    "MetricID": [random.choice(list(range(1, 501))) for _ in range(100)]  # Assuming MetricID ranges from 1 to 500
}
employees_df = pd.DataFrame(employees_data)

# Creating Sales DataFrame
sales_data = {
    "TotalSales": [round(random.uniform(1000, 10000), 2) for _ in range(500)],
    "SalesByCategory": [random.choice(categories_data["CategoryName"]) for _ in range(500)],
    "SalesByProduct": [random.choice(product_names) for _ in range(500)],
    "SalesByRegion": [fake.city() for _ in range(500)],
    "SalesByCity": [fake.city() for _ in range(500)]
}
sales_df = pd.DataFrame(sales_data)

# Creating Inventory Management DataFrame
inventory_management_data = {
    "ProductID": list(range(1, 501)),
    "CurrentStockLevels": [random.randint(1, 1000) for _ in range(500)],
    "InventoryTurnoverRate": [round(random.uniform(1, 10), 2) for _ in range(500)],
    "StockAlerts": ["Low" if random.random() < 0.1 else "Normal" for _ in range(500)]
}
inventory_management_df = pd.DataFrame(inventory_management_data)

# Creating Financial Metrics DataFrame
financial_metrics_data = {
    "Revenue": [round(random.uniform(10000, 100000), 2) for _ in range(500)],
    "Costs": [round(random.uniform(5000, 50000), 2) for _ in range(500)],
    "GrossMargin": [round(random.uniform(0.1, 0.5), 2) for _ in range(500)],
    "ProfitabilityByProductCategory": [random.choice(categories_data["CategoryName"]) for _ in range(500)]
}
financial_metrics_df = pd.DataFrame(financial_metrics_data)

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

#Exporting DataFrames to CSV files
products_df.to_csv('Products.csv', index=False)
categories_df.to_csv('Categories.csv', index=False)
customers_df.to_csv('Customers.csv', index=False)
orders_df.to_csv('Orders.csv', index=False)
employees_df.to_csv('Employees.csv', index=False)
sales_df.to_csv('Sales.csv', index=False)
inventory_management_df.to_csv('InventoryManagement.csv', index=False)
financial_metrics_df.to_csv('FinancialMetrics.csv', index=False)

# The generated CSV files can now be uploaded to Google Drive or any other storage service.
```
